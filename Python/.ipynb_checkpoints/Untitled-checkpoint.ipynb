```python
import cv2
import importlib
import numpy as np
from PIL import Image
```

```python
img = cv2.imread('mano.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
canvas = np.zeros(img.shape, np.uint8)
_, thresh = cv2.threshold(img, 25, 255, cv2.THRESH_BINARY)
im2_, contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

Image.fromarray(im2_)
```

```python
def max_area_contour(contours):
    cnt = contours[0]
    max_area = cv2.contourArea(cnt)

    for cont in contours:
        if cv2.contourArea(cont) > max_area:
            cnt = cont
            max_area = cv2.contourArea(cont)
    return cnt

cnt = max_area_contour(contours)

multiplier = 0.0001
perimeter = int(cv2.arcLength(cnt, True))
epsilon = multiplier * perimeter
approx = cv2.approxPolyDP(cnt, epsilon, True)  # parÃ¢metros para testar: epsilon(dita o quao simplificada fica a figura)
cv2.drawContours(canvas, [approx],  0, (50, 50, 50), 1)
aux = approx[-2:]
# cv2.drawContours(canvas, [aux],  0, (255, 255, 255), 1)
approx = np.concatenate((aux, approx), axis=0)

Image.fromarray(canvas)
```

```python
infl_p = np.diff(approx, n=2, axis=0)
infl_p
```

```python
sign = np.zeros(infl_p.shape[0])

for i in range(infl_p.shape[0]):
    if infl_p[i, 0, 1] == 0:
        if infl_p[i, 0, 1]/infl_p[i, 0, 0] > 0:
            sign[i] = 1
    else:
        if infl_p[i, 0, 0]/infl_p[i, 0, 1] > 0:
            sign[i] = 1
        
sign
```

```python
for i in range(sign.shape[0]):
    if sign[i-1] == 1:
        cv2.line(canvas, tuple(approx[i-1][0]), tuple(approx[i][0]), (255, 255, 255), 1)
        
Image.fromarray(canvas)
```
