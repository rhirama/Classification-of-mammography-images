```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pylab as plt
```

```python
pm_labels = np.arange(0.1, 2.1, 0.1)
pm_labels = np.round(pm_labels, 1)
```

```python
metric = 'acc'
clf = 'svm'

file = metric + '_p_value_matrix_' + clf + '.xlsx'
name = 'm' + metric + '_' + clf

excel = pd.ExcelFile(file)
p_values = excel.parse('Sheet1')

sb.heatmap(p_values, xticklabels=pm_labels, yticklabels=pm_labels, center=0.05, cmap=sb.dark_palette((1, 1, 1), n_colors=2), square=True, cbar=False)

plt.savefig(name, dpi=150)

plt.show()
```
